```python
from Bio import Entrez # the package biopython is needed(!)
```

```python
Entrez.email = 'your@email.com'

query="human gut microbiome"

handle = Entrez.esearch(db='pubmed', 
                        sort='relevance', 
                        retmax='20',
                        retmode='xml', 
                        term=query)

search_results = Entrez.read(handle)
```

```python
search_results
```

```python
# Show the list of found PubMed ids
search_results["IdList"]
```

```python
handle = Entrez.efetch(db='pubmed',
                       retmode='xml',
                       id=','.join(search_results["IdList"]))
article_metadata = Entrez.read(handle)
```
