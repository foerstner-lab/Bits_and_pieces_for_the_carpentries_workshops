# Metadaten über Econbiz beziehen
- [Econbiz Webseite](https://www.econbiz.de/)
- [Econbiz API Doku](https://api.econbiz.de/)

Die API eignet sich, um nach bestimmten Themen in Publikationen zu suchen.

```python
import requests
import json

# Base URL der EconBiz-API
base_url = "https://api.econbiz.de/v1/"

# Suchbegriff, nach dem du suchen möchtest (ersetze durch deinen eigenen Suchbegriff)
search_query = "knowledge management"

# Erstelle die vollständige URL für die Suchanfrage
full_url = f"{base_url}search?q={search_query}"

# Führe die HTTP-Anfrage an die API aus
response = requests.get(full_url)

dataset = json.loads(response.text)
```
