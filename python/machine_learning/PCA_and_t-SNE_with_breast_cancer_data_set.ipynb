```python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pca.fit(breast_cancer.data)
```

```python
pca_trans = pca.transform(breast_cancer.data)
```

```python
breast_cancer.data.shape
```

```python
pca_trans.shape
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.scatter(pca_trans[:, 0], pca_trans[:, 1], c=breast_cancer.target)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(breast_cancer.data)
```

```python
X_scaled = scaler.transform(breast_cancer.data)
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X_scaled)
```

```python
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=breast_cancer.target)
plt.legend(breast_cancer.target_names)
```

```python
plt.matshow(pca.components_)
plt.colorbar()
plt.xticks(range(len(breast_cancer.feature_names)),
breast_cancer.feature_names, rotation=60, ha='left')
```
