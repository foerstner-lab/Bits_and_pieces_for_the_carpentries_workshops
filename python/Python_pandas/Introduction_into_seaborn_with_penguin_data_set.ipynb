```python
import urllib.request
import pandas as pd
import seaborn as sns

# Downloading the data as saving it as local file
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
csv_file = "penguins.csv"
urllib.request.urlretrieve(url, csv_file)
```

```python
# Load the data into a pandas DataFrame
penguins = pd.read_csv(csv_file)
```

```python
# Show the DataFrame
penguins
```

### Using pandas simple in-built plotting capabilities

```python
# Generating a histogram of one column
penguins["body_mass_g"].hist()
```

```python
# Summary statistics for body mass of penguins
penguins["body_mass_g"].describe()
```

```python
# Increasing the granularity by highering the number of bins
penguins["body_mass_g"].hist(bins=20)
```

```python
# Histogram of penguins' bill lengths
penguins["bill_length_mm"].hist()
```

# Scatterplot

In the following we will encode an increasing number of attributes (numberical and categorial) by different aesthetics.

- positional x-axis
- positional y-axis
- area size
- color
- shape

```python
# Plotting two numerical attributes - one on the x axis, one on the y-axis

sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
# Now encode a third numerical attribute in the area size of the dots
# (this could now be called a bubble plot)

sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", size="body_mass_g")
```

```python
# Now enconding a fouth atttibute in the color value of the plot
sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                size="body_mass_g",
                hue="flipper_length_mm")
```

```python
# Now plotting two numerical attributes in the position 
# and one categorical attribut in the color.

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                hue="species")
```

```python
# Now encoding numerial attributes and one (another) categorical attribut in the color

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                hue="island")
```

```python
# Now plotting numerial attributes and on categorical attribut in the shape of the dots
# (harder to distiguis than color)

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                style="island")
```

```python
# Now plotting two numerial attributes in position, one categorial 
# attribute in color and one categorial attribute by shape

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                style="island",
                hue="sex")
```

```python
# Encoding two numerical values in the positions,
# one numerical in the size and two categorial 
# attributes in the color and in the shape.

# Warning - not necessarily a good idea

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                size="body_mass_g",
                style="island",
                hue="sex")
```

```python
# Not the categorial attribue "island" is encoded by shape and color; making
# recognizing the classes easier

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                size="body_mass_g",
                style="island",
                hue="island")
```

```python
# bad!
# Here the numerical attribute "body_mass_g" is encoded in the position on
# the y axis and the area size.
sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="body_mass_g",
                size="body_mass_g",
                style="island",
                hue="island")
```

## Distplots

```python
penguins.describe()
```

```python
# Box plot - only displaying some key statistical values
sns.boxplot(data=penguins, x="species", y="body_mass_g")
```

```python
# Box plot - as catplot
sns.catplot(data=penguins, kind="box", x="species", y="body_mass_g")
```

```python
# Swarm plot of body mass by species
sns.swarmplot(data=penguins, x="species", y="body_mass_g")
```

```python
# Horizontal violin plot: shows distribution of body mass along x-axis for each species on y-axis
sns.violinplot(data=penguins, x="species", y="body_mass_g")
```

```python
# Vertical violin plot: shows distribution of body mass along y-axis for each species on x-axis
sns.violinplot(data=penguins, y="species", x="body_mass_g")
```

```python
# Strip plot of body mass by species (each point represents an individual penguin)
sns.stripplot(data=penguins, x="species", y="body_mass_g")
```

```python
# Boxen plot of body mass by species
sns.boxenplot(data=penguins, x="species", y="body_mass_g")
```

```python
# Including another categorial attribute encoded as color.
sns.boxplot(data=penguins, x="species", y="body_mass_g", hue="sex")
```

```python
# Combining boxplot (via catplot which returns a FacetGrid object) and swarmplot

sns.catplot(data=penguins, kind="box", x="species", y="body_mass_g")
sns.swarmplot(data=penguins, x="species", y="body_mass_g", color="grey")
```

```python
# Combining boxplot and swarmplot

sns.catplot(data=penguins, kind="box", x="species", y="body_mass_g")
sns.stripplot(data=penguins, x="species", y="body_mass_g", color="grey")
```
