First, let's look at the data we will be working with. To do this, we'll go to GitHub: https://github.com/OpenAPC/openapc-de. 

APC stands for 'Article Processing Charges', which refers to the cost of publishing articles. OpenAPC is an organisation that aims to openly collect data showing how much German-speaking institutions pay for APCs. 

In the GitHub repository, there is a 'data' folder. In this folder, we find a folder called 'FU Berlin', which we will look at. Here, we have collected information for each year in CSV format. For example, we can look at the year 2015.

File name: [APC_FU_Berlin_2015.csv](https://github.com/OpenAPC/openapc-de/blob/master/data/fuberlin/APC_FU_Berlin_2015.csv).

As you can see, there is a table with different columns (institution, period, price in euros, Digital Object Identifier (DOI), hybrid journal, publisher, journal full title, etc.). Today, we will look at how much FU Berlin paid on average for APCs, and how the prices are distributed.

As we can see, the table is already nicely rendered in HTML by GitHub. If we click on 'Raw', we can see the 'raw' state of the data, as well as the comma separator and the fact that it is plain text. We can also see that the first row contains the columns names (institution, period, etc.). The second row is an entry.

Now, let's examine the data in the Jupyter notebook using Pandas. 

```python
import urllib.request # Module that defines functions and classes which help in opening URLs
import pandas as pd
```

```python
# Set the source URL
url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
apc_file = "openapc.csv"
```

```python
# Download the data and store them locally
urllib.request.urlretrieve(url, apc_file)
```

```python
# Read the CSV file into a Pandas dataframe
apcs = pd.read_csv(apc_file)
```

```python
# Display the content of the dataframe
# The Jupyter notebook recognises that the object is a dataframe and translates it into a nicely formatted table
apcs
```

```python
# Change the number of rows displayed to 101
pd.set_option("display.max_rows", 100)
apcs
```

```python
# Get the number of rows and columns
apcs.shape
```

```python
# Show the column names
apcs.columns
```

```python
# Select a single column
apcs["euro"]
```

```python
# Return the maximum value of a single column
apcs["euro"].max()
```

```python
# Return the minimum value of a single column
apcs["euro"].min()
```

```python
# Return the average value of a single column
apcs["euro"].mean()
```

```python
# Return the median value of a single column
apcs["euro"].median()
```

```python
# This alternative approach to handling the column is less secure, as it would fail if the column name contained spaces. 
# In general, spaces in column names should be avoided since computers interpret spaces as separators between commands.
apcs.euro
```

```python
# Identify cells greater or equal to a number
apcs["euro"] >= 2000
```

```python
# Show only articles with an APC of €2,000 or more
apcs[apcs["euro"] >= 2000]
```

✨ Exercise: Filtering Data
1. Filter all rows in whith the `"euro"` value is above the average.
2. Show only the `"journal_full_title"`, `"euro"` and `"publisher"`columns for these entries.

*Note:* to display multiple columns, use the following notation: `apcs[['"journal_full_title"', '"euro"', '"publisher"']]`

```python
# ✅ Solution
average_apc = apcs["euro"].mean() # Calculate average
over_average_apcs = apcs[apcs["euro"] > average_apc] # Apply filter
over_average_apcs[["journal_full_title", "euro", "publisher"]]
```

```python
# Plot a histogram showing the frequency of a particular APC in euros.  
apcs["euro"].hist()
```

```python
# Repeat, changing the bin size and colour
apcs["euro"].hist(bins = 20, color = "black") # bin = number of equal parts
```

```python
# Count the number of occurrences per publisher
apcs["publisher"].value_counts()
```

```python
# Plot this count
apcs["publisher"].value_counts().plot.bar()
```

```python
# Plot horizontally
apcs["publisher"].value_counts().plot.barh()
```
