## accumulate with groupby

```python
# save two columns to new dataframe
df_frl_file_data_small = df_frl_file_data[["isDescribedBy_createdBy", "hasData_size"]]
df_frl_file_data_small
```

```python
# sum "hasData_size" by "isDescribedBy_createdBy"
df_frl_file_data_small.groupby("isDescribedBy_createdBy").sum()
```

```python
# sum "hasData_size" by "isDescribedBy_createdBy" and sort descending
df_frl_file_data_small.groupby("isDescribedBy_createdBy").sum().sort_values(by="hasData_size", ascending=False)
```

```python
# plot horizontal pillar diagram for result of sum "hasData_size" by "isDescribedBy_createdBy" and sort ascending
matplotlib.pyplot.show(
    df_frl_file_data_small.groupby("isDescribedBy_createdBy").sum().sort_values(by="hasData_size", ascending=True).plot.barh()
)
```

## merge two dataframes

```python
# set url for second data file on github
url_parent = "https://raw.githubusercontent.com/lschruff/library_carpentry_workshops/refs/heads/main/2025-01-28_frl_parent_metadata_examples.csv"
```

```python
# load data directly into new dataframe
df_frl_parent_data = pd.read_csv(url_parent)
```

```python
# show head of new dataframe
df_frl_parent_data.head(20)
```

```python
# show column header of dataframe for parent data
df_frl_parent_data.keys()
```

```python
# show column header of dataframe for file data (as a reminder)
df_frl_file_data.keys()
```

```python
# we need a key column that holds the same name in each dataframe we want to merge
# in FRL, the IDs of the parents are written into the field "parentPid" on file level
# thus we can't just merge on "@id" as the IDs of these columns, although they have the same name, will never match up
# as a "quick and practical"-solution we will rename one of the columns of the parent data to match the file data
df_frl_parent_data.rename(columns={"@id": "parentPid"}, inplace=True)
df_frl_parent_data.keys()
```

```python
# now we merge the parent data onto the file data using column "parentPid" into new dataframe
df_frl_data_merged = df_frl_file_data.merge(df_frl_parent_data, how="left", on="parentPid")
```

```python
# show column headers of new dataframe
df_frl_data_merged.keys()
```

```python
# show new dataframe
df_frl_data_merged
```
