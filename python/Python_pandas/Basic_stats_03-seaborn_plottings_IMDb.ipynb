```python
import pandas as pd
import seaborn as sns
```

```python
imdb = pd.read_csv("https://datasets.imdbws.com/title.ratings.tsv.gz", sep="\t")
```

```python
imdb_subsample = imdb.sample(10000, random_state=1)
```

```python
sns.stripplot(imdb_subsample.averageRating)
```

```python
sns.stripplot(x=imdb_subsample.averageRating)
```

```python
sns.stripplot(y=imdb_subsample.averageRating)
```

```python
sns.stripplot(y=imdb_subsample.averageRating, alpha=0.1)
```

```python
sns.boxplot(y=imdb_subsample.averageRating)
```

```python
sns.boxenplot(y=imdb_subsample.averageRating)
```

```python
sns.violinplot(y=imdb_subsample.averageRating, alpha=0.1)
```

```python
sns.stripplot(y=imdb_subsample.numVotes, alpha=0.5)
```

```python
imdb_subsample["voting_class"] = imdb_subsample.numVotes.apply(
    lambda rating: "more_than_20" if rating > 20 else "less_or_equal_than_20"
)
```

```python
imdb_subsample
```

```python
imdb_subsample.voting_class.value_counts()
```

```python
sns.stripplot(imdb_subsample[imdb_subsample.voting_class == "more_than_20"].averageRating)
```

```python
sns.stripplot(imdb_subsample[imdb_subsample.voting_class == "less_or_equal_than_20"].averageRating)
```

```python
sns.stripplot(y=imdb_subsample.averageRating, x=imdb_subsample.voting_class)
```

```python
sns.stripplot(y=imdb_subsample.averageRating, x=imdb_subsample.voting_class, alpha=0.5)
```

```python
sns.violinplot(y=imdb_subsample.averageRating, x=imdb_subsample.voting_class)
```

```python
sns.boxplot(y=imdb_subsample.averageRating, x=imdb_subsample.voting_class)
```

```python
sns.boxenplot(y=imdb_subsample.averageRating, x=imdb_subsample.voting_class)
```
