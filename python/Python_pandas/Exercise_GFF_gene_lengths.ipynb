## Solution:

```python
%matplotlib inline
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
for gff_file, source_url in gff_files_and_source_urls.items():
    urllib.request.urlretrieve(source_url, gff_file)
    gff_data = pd.read_csv(gff_file, sep="\t", comment="#", 
                       names=["seqid", "source", "type", "start", "end", "score", "strand", "phase", "attributes"])
    gff_data["length"] = gff_data["end"] - gff_data["start"] + 1
    gene_data = gff_data[gff_data["type"] == "gene"]
    gene_data["length"].hist(bins=30, color="black")
    plt.show()
```

## Bonus task: Add the species name to the histogram

## Solution:

```python
%matplotlib inline
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
for gff_file, source_url in gff_files_and_source_urls.items():
    urllib.request.urlretrieve(source_url, gff_file)
    gff_data = pd.read_csv(gff_file, sep="\t", comment="#", 
                       names=["seqid", "source", "type", "start", "end", "score", "strand", "phase", "attributes"])
    gff_data["length"] = gff_data["end"] - gff_data["start"]
    gene_data = gff_data[gff_data["type"] == "gene"]
    gene_data["length"].hist(bins=30, color="black")
    plt.show()
    species = gff_file.split(".")[0]
    print(f"gene lengths of {species}")
```

## Solution:

```python
%matplotlib inline
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
for gff_file, source_url in gff_files_and_source_urls.items():
    urllib.request.urlretrieve(source_url, gff_file)
    gff_data = pd.read_csv(gff_file, sep="\t", comment="#", 
                       names=["seqid", "source", "type", "start", "end", "score", "strand", "phase", "attributes"])
    gff_data["length"] = gff_data["end"] - gff_data["start"]
    gene_data = gff_data[gff_data["type"] == "gene"]
    gene_data["length"].hist(bins=30, color="black")
    plt.show()
    species = gff_file.split(".")[0].replace("_", ". ")
    print(f"           gene lengths of {species}")
```
